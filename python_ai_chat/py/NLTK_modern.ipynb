## 1-2. 문장 테이블 만들기 [STEM version]

### 아래부터 각 문장들 tokenizing table 생성(Actual)

```python
with open('modern_s1_e6.txt', 'r', encoding='utf8') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
data_list = [x.strip().lower() for x in content]
print(data_list)
```

```python
# NLTK and arrays
from nltk.tokenize import sent_tokenize, word_tokenize
# phrases = sent_tokenize(data_list)

# 영어만 추리기 위한 패키지
from nltk.corpus import wordnet

# 특수기호 제거 패키지
import re
```

```python
# 문장 테이블 만들기
words_table = word_dic_table.copy()

for i in range(len(data_list)):
    
    # 특수기호 삭제
    data_list_clean = re.sub('[^A-Za-z0-9]+', ' ', data_list[i])
    
    
    # 각 문장 tokenizing
    words_list = word_tokenize(data_list_clean)

    # 각 문장 Stemming
    ps = PorterStemmer()
    words_stemmed = []
    for word in words_list:
        stem_word = ps.stem(word)
        # print(stem_word)
        words_stemmed.append(stem_word)
    
    
    # 영어 단어만 추림
    # from nltk.corpus import wordnet
    word_eng = []

    for word in words_stemmed:
        if not wordnet.synsets(word):
          #Not an English Word
            continue
        else:        
            # English Word
            word_eng.append(word)
            # print(word)     
        

    words_list = word_eng

    # 각 문장의 단어들이 사전에 있는지 확인
    matching = [s for s in words_list if s in word_dic_list]

    # 문장별 테이블 생성
    matching = pd.Series(matching).value_counts()
    matching_df = pd.DataFrame({i:matching.values}, index=matching.index)

    # 문장 테이블에 추가
    words_table = words_table.join(matching_df, how='outer')

# NAN 값을 0으로 변경
words_table.fillna(0, inplace=True)

# float 값을 정수형을 변경
words_table_int = words_table.astype('int64')
```

```python
def sentence_search(input_sentence):

    # 문장 테이블 콘테이너
    words_table  = word_dic_table.copy()    # 문장 테이블 초가화 폼                               

    # 특수기호 삭제
    input_sentence = re.sub('[^A-Za-z0-9]+', ' ', input_sentence)


    # input 문장 tokenizing
    words_list = word_tokenize(input_sentence)      # not stemmed
    # print(words_list)

    # input 문장 Stemming
    ps = PorterStemmer()
    words_stemmed = []

    for word in words_list:
            stem_word = ps.stem(word)
            # print(stem_word)
            words_stemmed.append(stem_word)      # stemmed
    # print(words_stemmed)


    # 영어 단어만 추림
    # from nltk.corpus import wordnet
    word_eng = []

    for word in words_stemmed:
        if not wordnet.synsets(word):
          #Not an English Word
            continue
        else:        
            # English Word
            word_eng.append(word)
            # print(word)     


    words_list = word_eng


    # input 문장의 단어들이 사전에 있는지 확인
    matching = [s for s in words_list if s in word_dic_list]
    # print(matching)
    matching = pd.Series(matching).value_counts()
    # print(matching)
    matching_df = pd.DataFrame({'input_sentence':matching.values}, index=matching.index)
    # print(matching_df)

    # 문장 테이블에 추가
    words_table = words_table.join(matching_df, how='outer')

    # NAN 값을 0으로 변경
    words_table.fillna(0, inplace=True)

    # float 값을 정수형을 변경
    words_table_input = words_table.astype('int64')
    
    
    
    ## 문장간 유클리도 거리 계산
    sentence_distance = []

    words_table_input_list = list(words_table_input['input_sentence'].values)  # 입력 문장의 단어 

    for i in range(words_table_int.shape[1]):
        words_table_sentence_list = list(words_table_int[i].values)    # 문장 테이블의 단어 : i 순번
        distance_each = euclidean_distance(words_table_input_list, words_table_sentence_list)
        # print(distance_each)
        sentence_distance.append(distance_each)

    sentence_like_index =  sentence_distance.index(min(sentence_distance)) # 가장 가까운 거리 문장
   
    
    
    ### return 값(대답): 가장 비슷한 문장의 다음 문장
    # 가장 비슷한 문장
    data_sentence_list[sentence_like_index]
    
    # 대답 : 그 다음 문장
    return(data_sentence_list[sentence_like_index+1])
    
```
